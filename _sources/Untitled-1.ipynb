## **CONSTRUCCIÓN DE ÁRBOLES DE DECISIÓN CON GINI**

Primeramente, veamos cuáles son las fórmulas necesarias para la realización de este ejercicio.

1.  **índice Gini para cada categoría de una característica**

$$G = 1 - \sum_{i=1}^C p_i^2$$

donde: 

- C corresponde al número de clases en el nodo.
​
- $P_i$ corresponde a la proporción de ejemplos pertenecientes a la clase. 

2. **Índice Gini total de una característica**

$$G_{\text{split}} = \sum_{j=1}^m \frac{N_j}{n} G_j$$

- $N_j$: Número de ejemplos en la partición $j$.
- $N$: Número total de ejemplos en el conjunto de datos.
- $G_j$: Índice de Gini de la categoría $j$.
